```python
import torch
import torch.nn as nn
```

```python
batch, sequence_length, embedding_dim = 20, 5, 10
embedding = torch.randn(batch, sequence_length, embedding_dim)
embedding
```

```python
first_setence_embedding = torch.concat([torch.ones(1, embedding_dim) * i for i in range(1, sequence_length + 1)])
first_setence_embedding
```

```python
embedding[0] = first_setence_embedding
```

```python
layer_norm = nn.LayerNorm((5,10))
normed_embedding = layer_norm(embedding)
```
